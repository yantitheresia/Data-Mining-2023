```python
# Mempersiapkan library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
```

```python
# Memanggil Dataset
dataset = pd.read_excel('Data Keluar.xlsx')
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, 1].values
```

```python
dataset.keys()
```

```python
dataset.shape
```

```python
#Menampilkan isi sebagian dataset
dataku = pd.DataFrame(dataset)
dataku.head()
```

```python
# Split dataset menjadi Set dan Testing Set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state=0)
```

```python
# Melakukan Fitting Simple Linear Regression pada Training Set
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
# Memprediksi Data
plt.scatter(dataku.BiayaKeluar, dataku.NilaiBiaya)
plt.xlabel("Biaya Keluar")
plt.ylabel("Nilai Biaya")
plt.title("Grafik Biaya Keluar vs Nilai Biaya")
plt.show
```

```python
# Visualisasi Hasil Prediksi Pada Training-Set
# Ukuran Plot
plt.figure(figsize=(10,8))

# Biru adalah data observasi
plt.scatter(x_train, y_train, color = 'blue')

# Garis merah adalah hasil prediksi dari machine learning
plt.plot(x_train, regressor.predict(x_train), color = 'red')

# Memberi Judul dan Label
plt.title('Biaya Keluar terhadap Nilai Biaya (Training Set)')
plt.xlabel('Biaya Keluar')
plt.ylabel('Nilai Biaya')

# Menampilkan plot
plt.show()
```

```python
# Visualisasi Hasil Prediksi Pada Test-Set
# Ukuran plot
plt.figure(figsize=(10,8))

# Biru adalah data observasi
plt.scatter(x_test, y_test, color = 'blue')

# Garis merah adalah hasil hasil prediksi dari me=achine learning
plt.plot(x_test, regressor.predict(x_test), color = 'red')

# Memberi Judul dan Label
plt.title('Biaya Keluar terhadap Nilai Biaya (Test Set)')
plt.xlabel('Biaya Keluar')
plt.ylabel('Nilai Biaya')

# Menampilkan plot
plt.show
```
